# Part 1

```python
def check_neighbors( pt, grid):
    r, c = pt
    cnt = 0 

    if grid[r][c] =='.':
        return float('inf')
    
    for ir in [-1,0,1]:
        nr = ir+r
        if nr >= 0 and nr < len(grid):
            for ic in [-1,0,1]:
                if ir == ic == 0:
                    continue
                nc = ic+c
                
                if nc >= 0  and nc < len(grid[0]):
                    cnt+=int(grid[nr][nc] =='@' )
                    

    return cnt
        
def count_accessible(grid):
    cnt=0
    for r, row in enumerate(grid):
        for c, ele in enumerate(row):
            cnt+=int( check_neighbors((r,c), grid) <4)
            
    return cnt
```

```python
def part_one(file_path='input_data/test_04.txt'):
    grid=[]
    with open(file_path,'r') as f:
        for line in f.readlines():
            grid.append( list(line.strip()))
    return count_accessible(grid)
           
```

# Part 2


```python
def do_one_round(grid):
    change_set = []
    for r, row in enumerate(grid):
        for c, ele in enumerate(row):
            if check_neighbors((r,c), grid) <4:
                change_set.append( (r,c))

    if len(change_set) > 0:
        for pt in change_set:
            r,c = pt
            grid[r][c] ='.'
        return do_one_round(grid)

    return (''.join( [ ''.join(g) for g in grid])).count('@')
```

```python
def part_two(file_path='input_data/test_04.txt'):
    grid=[]
    with open(file_path,'r') as f:
        for line in f.readlines():
            grid.append( list(line.strip()))

    original_rolls = (''.join( [ ''.join(g) for g in grid])).count('@')
    leftover = do_one_round(grid)
    return original_rolls - leftover
           
```

```python
%%time
part_two()
```

```python
%%time
part_two('input_data/day_04.txt')
```
