# Part 1

```python
def find_splits(grid):
    split_locs = set([])

    for r in range(1,len(grid)):
        prev_row = grid[r-1]

        for c, ele in enumerate(prev_row):
            if ele in ['|', 'S']:
                if grid[r][c] == '^':
                    split_locs.add((r,c))
                    if c-1 >= 0 :
                        grid[r][c-1] = '|'
                    if c+1 < len(prev_row) :
                        grid[r][c+1] = '|'
                else:
                    grid[r][c] = '|'
    return len(split_locs)
                    
    
def part_one(file_path='input_data/test_07.txt'):
    grid =[]
    
    with open(file_path,'r') as f:
        for line in f.readlines():
            grid.append( list(line.strip()) )

    

    return find_splits(grid)
           
```

# Part 2


```python
from collections import defaultdict

def find_splits_timeline(grid):
    split_locs = set([])

    beam_dict=defaultdict(int)
    for c,e in enumerate(grid[0]):
        if e=="S":
            beam_dict[(0,c)]=1

    for r in range(1,len(grid)):
        prev_row = grid[r-1]

        for c, ele in enumerate(prev_row):
            if ele in ['|', 'S']:
                if grid[r][c] == '^':
                    split_locs.add((r,c))
                    if c-1 >= 0 :
                        grid[r][c-1] = '|'
                        beam_dict[(r,c-1)]+=beam_dict[(r-1,c)]
                    if c+1 < len(prev_row) :
                        grid[r][c+1] = '|'
                        beam_dict[(r,c+1)]+=beam_dict[(r-1,c)]
                else:
                    grid[r][c] = '|'
                    beam_dict[(r,c)]+=beam_dict[(r-1,c)]

    timelines = sum([ v for k,v in beam_dict.items() if k[0] == len(grid)-1])
                    
    return len(split_locs), timelines
                    
    
def part_two(file_path='input_data/test_07.txt'):
    grid =[]
    
    with open(file_path,'r') as f:
        for line in f.readlines():
            grid.append( list(line.strip()) )

    
    return find_splits_timeline(grid)
           
```

```python
%%time
part_two()
```

```python
%%time
part_two('input_data/day_07.txt')
```
