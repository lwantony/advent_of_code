# Part 1

```python
def part_one(file_path='input_data/test_01.txt'):
    state=50
    nzero=0
    with open(file_path,'r') as f:
        for line in f.readlines():
            turn = line.strip().replace('L','-').replace('R','')
            turn = int(turn)
            state = (state+turn)%100
            if state == 0:
                nzero+=1
    
    return nzero
```

```python
def part_two(file_path='input_data/test_01.txt'):
    state=50
    nzero=0
    pass_zero =0
    with open(file_path,'r') as f:
        for line in f.readlines():
            turn = line.strip().replace('L','-').replace('R','')
            turn = int(turn)
            
            full_turns =abs(turn)//100
            turn = turn - full_turns*100 if turn >0 else turn + full_turns*100

            prior_state=state
            state += turn
            if state > 100 or (prior_state !=0 and state<0) :
                pass_zero+=1

            pass_zero+= full_turns
            
            state = state%100    
            if state == 0:
                nzero+=1
    
    return nzero,pass_zero +nzero
```

```python
part_two()
```

```python
part_two('input_data/day_01.txt')
```
