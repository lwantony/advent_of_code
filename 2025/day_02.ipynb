```python
import numpy as np
```

```python
def find_all_repeat(lower, upper):
    lower_mag = int(np.floor(np.log10(lower)))
    upper_mag = int(np.floor(np.log10(upper)))

    count = 0
    total = 0 
    seen = set([])

    for mag in range(lower_mag, upper_mag+1):
        cap_repeat = int((mag+1)//2)
        
        for n_repeat in range(1,cap_repeat+1):
            if (mag+1)%n_repeat != 0 :
                continue
            n_units = int((mag+1)/n_repeat)
            
            unit_lower = int('1'+'0'*(n_repeat-1))
            unit_upper = int('9'*(int(n_repeat)))
            for unit in range(unit_lower, unit_upper+1):
                full_number = int(str(unit)*n_units)
                
                if full_number > upper :
                    break
                if full_number >= lower and full_number <= upper :
                    if full_number not in seen:
                        count +=1
                        total += full_number
                        seen.add(full_number)
    return count, total
```

```python
def find_invalid_number(lower, upper):
    lower_mag = np.floor(np.log10(lower))
    upper_mag = np.floor(np.log10(upper))

    count = 0
    part_one = 0 
    
    if lower_mag == upper_mag and lower_mag%2 == 0:
        return 0, 0

    if upper_mag-lower_mag == 1 and lower_mag%2 == 0: 
        lower = int(10**upper_mag)
        lower_mag = upper_mag
        
    elif upper_mag-lower_mag == 1 and lower_mag%2 == 1: 
        upper = int('9'*(int(lower_mag)+1))
        upper_mag = lower_mag
        
    if lower_mag == upper_mag and lower_mag%2 == 1:
        low_str = str(lower)
        up_str = str(upper)

        str_len = int(len(low_str)/2)
        start = low_str[:str_len]
        end= up_str[:str_len]
        for num in np.arange(int(start),int(end)+1):
            check = int(str(num)+str(num))
            if check >= lower  and check <= upper:
                count+=1
                part_one +=check
            if check > upper:
                return count, part_one
                
    if upper_mag-lower_mag > 1:
        for mag in range(lower_mag, upper_mag+1):
            if mag%2 == 0:
                continue
            if mag > lower_mag and mag < upper_mag:
                tmp_upper = int('9'*(int(mag)+1))
                tmp_lower = int(10**mag)
                tmp_cnt, tmp_pt1= find_invalid_number(tmp_upper,tmp_upper)
                
                count += tmp_cnt
                part_one+= tmp_pt1
                
            elif mag == lower_mag:
                tmp_upper = int('9'*(int(mag)+1))
                tmp_cnt, tmp_pt1= find_invalid_number(lower,tmp_upper)
                count += tmp_cnt
                part_one+= tmp_pt1
                
            elif mag == upper_mag:
                tmp_lower = int(10**upper_mag)
                tmp_cnt, tmp_pt1= find_invalid_number(tmp_lower,upper)
                count += tmp_cnt
                part_one+= tmp_pt1
    return count,part_one
            

class Intervals:
    def __init__(self,lower,upper):
        self.lower = lower
        self.upper = upper
        cnt, pt1= find_invalid_number(lower,upper)
        self.n_invalid = cnt
        self.part_one = pt1

    def part_two(self):
        return find_all_repeat( self.lower, self.upper)
        
```

```python
def part_one(file_path='input_data/test_02.txt'):
    ints=[]
    
    with open(file_path,'r') as f:
        for line in f.readlines():
            for i in line.strip().split(','):
                low,up = i.split('-')
                ints.append(Intervals(int(low),int(up)))
                
    return sum([ i.part_one for i in ints])
```

```python
def part_two(file_path='input_data/test_02.txt'):
    ints=[]
    
    with open(file_path,'r') as f:
        for line in f.readlines():
            for i in line.strip().split(','):
                low,up = i.split('-')
                ints.append(Intervals(int(low),int(up)))
                
    return sum([ i.part_two()[1] for i in ints])
```
