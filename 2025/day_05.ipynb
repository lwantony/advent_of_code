```python
class Interval:
    def __init__(self, lower, upper):
        self.lower= int(lower)
        self.upper = int(upper)

    def __lt__(self, other):
        return self.lower <= other.lower

    def in_range(self, num):
        return self.lower <= num and num <= self.upper 
        
    def overlap(self, other):
        if other.lower <= self.upper and self.lower <= other.upper:
            return True
        return False
        
    def numbers_in_range(self):
        return self.upper - self.lower + 1
```

```python
def part_one(file_path='input_data/test_05.txt'):
    ints =[]
    get_intervals = True 
    cnt=0
    with open(file_path,'r') as f:
        for line in f.readlines():
            line = line.strip()
            if  len(line) < 1:
                get_intervals = False
                ints.sort()
                continue
                
            if get_intervals:
                l,u = line.split('-')
                ints.append( Interval(l,u))
                
            else:
                number = int(line)

                for it in ints:
                    if it.in_range(number):
                        cnt+=1
                        break
         
    return cnt
           
```

```python
%%time
part_one()
```

```python
def combine_ints(intervals):
    intervals.sort()
    new_ints = []
    cur_int = intervals.pop(0)
    for next_int in intervals:
        if cur_int.overlap(next_int):
            cur_int = Interval( min(cur_int.lower,next_int.lower),
                                max(cur_int.upper,next_int.upper)
                              )
        else:
            new_ints.append(cur_int)
            cur_int = next_int
    new_ints.append(cur_int)
    return new_ints
    
```

```python
def part_two(file_path='input_data/test_05.txt'):
    ints =[]
    get_intervals = True 
    cnt=0
    with open(file_path,'r') as f:
        for line in f.readlines():
            line = line.strip()
            if  len(line) < 1:
                get_intervals = False
                ints.sort()
                continue
                
            if get_intervals:
                l,u = line.split('-')
                ints.append( Interval(l,u))
                
            else:
                number = int(line)

                for it in ints:
                    if it.in_range(number):
                        cnt+=1
                        break
                        
    reduced_ints = combine_ints(ints)

    total = sum([ i.numbers_in_range() for i in reduced_ints])
         
    return cnt,total
           
```
