```python
def find_best_jolt(number_string,num_len=12):
    num_list = list(number_string)
    list_len = len(num_list)

    start=0
    number=''
    for i in range(num_len-1,-1,-1):
        max_idx = num_list[start:list_len-i].index(max(num_list[start:list_len-i]))
        number+=num_list[start:list_len-i][max_idx]
        start = start+max_idx+1
    return int(number)
```

```python
find_best_jolt('234234234234278')
```

```python
def part_one(file_path='input_data/test_03.txt'):
    total=0 
    tot_12 = 0
    with open(file_path,'r') as f:
        for line in f.readlines():
            total+=find_best_jolt(line.strip(),2)
            tot_12+=find_best_jolt(line.strip())
    return total, tot_12
```
