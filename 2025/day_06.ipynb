# Part 1

```python
import math
operations={ '+': sum,
             '*': math.prod}

def part_one(file_path='input_data/test_06.txt'):
    grid =[]
    with open(file_path,'r') as f:
        for line in f.readlines():
            grid.append( line.strip().split() )
    ops = grid.pop()

    total = 0 
    for c,op in enumerate(ops):
        numbers = [ int(r[c]) for r in grid if c < len(r) and r[c].isdigit()]
        total += operations[op](numbers)
        
    return total
           
```

```python
%%time
part_one()
```

```python
%%time
part_one('input_data/day_06.txt')
        
```

```python
import math
operations={ '+': sum,
             '*': math.prod}

def part_two(file_path='input_data/test_06.txt'):
    grid =[]
    with open(file_path,'r') as f:
        for line in f.readlines():
            grid.append( list(line.strip('\n') ))
            
    ops = grid.pop()

    op_points = [ c for c,ele in enumerate(ops) if ele !=' ']

    
    op_points.append(max( [ len(g) for g in grid])+1)

    total = 0 
    for p,(e,s) in enumerate(zip(op_points, op_points[1:])):
        numbers = []
        for c in range(s-2, e-1, -1):
            num=''
            for r in grid:
                if len(r) > c:
                    num = num+r[c]
            if num.strip().isdigit():
                numbers.append(int(num))


        op = ops[e]
        total += operations[op](numbers)
    
    return total
           
```

```python
%%time
part_two()
```

```python
%%time
part_two('input_data/day_06.txt')
```
