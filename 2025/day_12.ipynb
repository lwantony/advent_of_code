```python
from hmac import new
from socket import if_indextoname
import numpy as np
from queue import PriorityQueue
from scipy.signal import convolve2d
import itertools



# np.rot90 rotation counter clockwise
# np.fliplr will flip the array left to right

class Presents:
    def parse_files(self,file):

        presents = []
        trees = []

        with open(file,'r') as f:
            line = f.readline()
            while line:
                if ":" in line:
                    inp, out = line.strip().split(':')
                    if 'x' in inp:
                        trees.append( ([ int(t) for t in inp.split('x') ],
                                       [ int(p) for p in out.strip().split() ])
                                       )
                    else:
                        present =[]
                        line = f.readline()
                        while line != '\n':
                            present.append([ int(c == '#') for c in list(line.strip()) ])
                            line = f.readline()
                        presents.append(np.vstack(present))
                line = f.readline()
        return presents, trees

    def present_rotations(self,pres_idx):
        pres = self.presents[pres_idx]
        mats = []
        seen = set()

        for k in range(4):
            r = np.rot90(pres, k)
            key = r.tobytes()
            if key not in seen:
                seen.add(key)
                mats.append(r)
                mats.append(np.fliplr(r))
        return mats
    
    def check_tree(self,tree_index):
        dim, pres_cnt = self.trees[tree_index]
        floor = np.ones(dim)
        frontier = PriorityQueue()
        counter = itertools.count()
        frontier.put((sum(pres_cnt), ( next(counter),tuple(pres_cnt),floor)))
        seen=set([])

        while not frontier.empty():
            cnt,state = frontier.get()
            _, pres_cnt,floor = state

            pres = np.nonzero(pres_cnt)[0][0]

            orients = self.present_rotations(pres)

            fit_cnt = self.presents[pres].sum()
            for o in orients:
                check = convolve2d(floor,o,mode='valid')
                for rshift, row in enumerate(check):
                    for yshift, val in enumerate(row):
                        if val == fit_cnt:
                            new_floor = floor.copy()
                            new_cnt = np.array(pres_cnt).copy()
                            new_cnt[pres] -=1

                            for r, prow in enumerate(o):
                                for c, pval in enumerate(prow):
                                    new_floor[rshift+r][yshift+c] -= pval
                            if sum(new_cnt) == 0:
                                return True

                            key=(new_floor.tobytes(),new_cnt.tobytes())
                            if key not in seen:
                                seen.add(key)
                                frontier.put((sum(new_cnt), (next(counter),tuple(new_cnt),new_floor)))
        return False





        
    def __init__(self,file_path):

        self.presents, self.trees = self.parse_files(file_path) 
                
```
